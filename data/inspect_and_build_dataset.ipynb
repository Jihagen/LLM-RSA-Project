## 4. Verify: check every word appears in every sentence

```python
issues = []
for word, senses in PROPOSED_DATA.items():
    for sid, (label, sents) in senses.items():
        for i, s in enumerate(sents):
            if word.lower() not in s.lower():
                issues.append(f"  MISSING '{word}' in sense-{sid} sentence {i+1}: {s}")

if issues:
    print("⚠️  Problems found:")
    for msg in issues:
        print(msg)
else:
    print("✅  All sentences contain the target word.")

# Summary
total = sum(len(sents) for senses in PROPOSED_DATA.values() for _, sents in senses.values())
print(f"\nTotal sentences: {total}  ({len(PROPOSED_DATA)} homonyms × 2 senses)")
```
